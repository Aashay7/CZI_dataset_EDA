```python
FILE_LOCATION = "data_original/all_dataset_summaries"

organ_dict = {}
organ_ctr = {}
unique_organ_list = ['kidney', 'lung', 'heart', 'blood', 'brain', 'pancreas', 'fetal', 'lung']
#counter_dict = defaultdict(unique_organ_list)
for filename in Path(FILE_LOCATION).glob('*'):
    fname = str(filename).split('/')[-1]
    #print(f"{filename}")
    data = pd.read_csv(filename)
    for organ in list(data['tissue_type'].unique()):
        if organ not in organ_ctr.keys():
            organ_ctr[organ]=0
        organ_ctr[organ]+=1
    
    organ_dict[fname] = {
        'dataset_name' : list(data['dataset_name'].unique())[0],
        'unique_organs' : list(data['tissue_type'].unique())
    }

    
```

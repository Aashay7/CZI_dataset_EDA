```python
import numpy as np
from pathlib import Path
import pandas as pd
import json
from collections import defaultdict
import requests
from tqdm import tqdm

pd.set_option("max_columns", None)
pd.set_option("max_rows", None)
```

```python
df = pd.read_csv("data_original/master_table.csv")
df.head()
```

```python
def get_doi_data(dataset_name):
    parameters = {"rows": 5, "query.title": dataset_name}
    endpoint_url = f"https://api.crossref.org/works"
    response = requests.get(endpoint_url, params=parameters)
    title_is_same = False
    try:
        response.raise_for_status()
        response_json = response.json()
        title_is_same = True if response_json["message"]["items"][0]["title"][0].strip(".").lower() == dataset_name.strip(".").lower() else False
        total_results = int(response_json["message"]["total-results"])
    except Exception as e:
        print("Error : ", e)
    doi_data = {
        "doi": response_json["message"]["items"][0]["DOI"],
        "dataset_name": dataset_name,
        "title": response_json["message"]["items"][0]["title"][0],
        "url": response_json["message"]["items"][0]["URL"],
        "title_is_same": title_is_same,
        "total-results": total_results,
        }
    print(json.dumps(response_json, indent=4))

    return doi_data
```

```python
doi_df = pd.DataFrame(index=None, columns=["tissue_type", "doi", "dataset_name", "title", "url", "title_is_same", "total-results"])

for dataset_name in tqdm(df["dataset_name"].unique()):
    doi_df.loc[len(doi_df)] = [df[(df['dataset_name'] == dataset_name)].reset_index().loc[0]['tissue_type'], *get_doi_data(dataset_name).values()]
```
